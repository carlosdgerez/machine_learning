```
import altair as alt
points = alt.Chart(housing.head(5000)).mark_circle().encode(
    longitude='long:Q',
    latitude ='lat:Q',
    size= alt.value(10),
    tooltip='new_date'
).project('albersUsa').properties (width =500,
                                    height= 400)
```

```
from vega_datasets import data

states = alt.topo_feature(data.us_10m.url, feature='states')

map = alt.Chart(states).mark_geoshape(
    fill='lightgray',
    stroke='white'
).project('albersUsa').properties(
    width=500,
    height=300
)
```

```
locations = (map + points)
locations
```
